```python
# 1. Load dataset
df = pd.read_csv("empsalary.csv")

```

```python
# Check for missing values
print(df.isnull().sum())

# Drop duplicates
df = df.drop_duplicates()

# Fill or drop missing values (example: drop rows with missing values)
df = df.dropna()
```

```python
# 2. Clean data
df.replace(' ?', np.nan, inplace=True)
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

```

```python
print(df.columns)
```

```python
# exploratory data analysis check balance
#countplot
sns.countplot(x='income', data=df)
plt.title('Income Distribution')
plt.show()
```

```python
#Correlation heatmap
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Now compute correlation and plot
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 4. Encode categorical columns
categorical_cols = df.select_dtypes(include='object').columns.tolist()
if 'income' in categorical_cols:
    categorical_cols.remove('income')

le_dict = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    joblib.dump(le, f"{col}_encoder.pkl")  # Save encoder for Streamlit
    le_dict[col] = le

```

```python
# 5. Encode target column
if df['income'].dtype == 'object':
    target_le = LabelEncoder()
    df['income'] = target_le.fit_transform(df['income'])
    joblib.dump(target_le, "target_encoder.pkl")

```

```python
# 6. Feature scaling
X = df.drop("income", axis=1)
y = df["income"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, "scaler.pkl")

# Save feature column names
joblib.dump(X.columns.tolist(), "model_columns.pkl")

```

```python
# 7. Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

```

```python
# 8. Model training
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"\n{name} Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(classification_report(y_test, y_pred))

```

```python
# 9. Save best model (Random Forest here)
best_model = RandomForestClassifier()
best_model.fit(X_scaled, y)
joblib.dump(best_model, "best_model.pkl")

print(" All files saved: encoders, scaler, model, column names")
```
